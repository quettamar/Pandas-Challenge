```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data.nunique()
players = total_players["SN"]
players_df = pd.DataFrame([[total_players["SN"]]], 
columns = ["Number of Total Players"])
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].value_counts())
average_price = purchase_data["Price"].sum() / len(purchase_data)
number_of_purchases = len(purchase_data)
total_revenue = purchase_data["Price"].sum()
unique_items
average_price
number_of_purchases
total_revenue
purchasing_df = pd.DataFrame([[unique_items, average_price, number_of_purchases, total_revenue]], 
columns = ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"])
purchasing_df["Average Price"] = purchasing_df["Average Price"].astype(float).map("${:,.2f}".format)
purchasing_df["Total Revenue"] = purchasing_df["Total Revenue"].astype(float).map("${:,.2f}".format)
purchasing_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df = purchase_data[['SN', 'Gender']]
gender_count_df = purchase_data.drop_duplicates(subset="SN", keep="first", inplace=False)

#creating variables that define the different gender populations
male_pop_df = gender_count_df.loc[gender_count_df["Gender"] == "Male"]
men = len(male_pop_df)
male_percentage = (men / players) 

female_pop_df = gender_count_df.loc[gender_count_df["Gender"] == "Female"]
women = len(female_pop_df)
female_percentage = (women / players)

other_pop_df = gender_count_df.loc[gender_count_df["Gender"] == "Other / Non-Disclosed"]
other_pop = len(other_pop_df)
other_percentage = (other_pop / players)

gender_demographics_df = pd.DataFrame([
    {" " : "Male", "Total Count" : men, "Percentage of Players" : male_percentage},
    {" " : "Female", "Total Count" : women, "Percentage of Players" : female_percentage},
    {" " : "Other / Non-Disclosed", "Total Count" : other_pop, "Percentage of Players" : other_percentage}
    ]).set_index(" ")

gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].astype(float).map("{:,.2%}".format)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_mean = gender_count_df.groupby(["Gender"])
gender_mean_df = gender_mean.mean()
gender_groupcount = purchase_data.groupby(["Gender"])
gender_groupcount_df = gender_groupcount.sum()
gender_average = purchase_data.groupby(["Gender"])
gender_average_df = gender_average.mean()

```

```python

gender_mean_df["Purchase Count"] = purchase_data["Gender"].value_counts()
gender_mean_df = gender_mean_df.drop(columns=["Purchase ID", "Age", "Item ID", "Price"])
gender_mean_df["Average Purchase Price"] = gender_average_df["Price"]
gender_mean_df["Total Purchase Value"] = gender_groupcount_df["Price"]
gender_mean_df["Avg Total Purchase per Person"] =  gender_groupcount_df["Price"] / gender_demographics_df["Total Count"]

```

```python
gender_mean_df["Average Purchase Price"] = gender_mean_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_mean_df["Avg Total Purchase per Person"] = gender_mean_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
gender_mean_df["Total Purchase Value"] = gender_mean_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
gender_mean_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


gender_count_df["Age Group"] = pd.cut(gender_count_df["Age"], bins, labels = labels)
age_demographic = gender_count_df.groupby(["Age Group"])
age_demographic_df = age_demographic.count()
age_demographic_df = age_demographic_df.rename(columns={"SN": "Total Count"})
age_demographic_df = age_demographic_df.drop(columns=["Purchase ID", "Age", "Item ID", "Gender", "Item Name", "Price"])
age_demographic_df["Percentage of Players"] = age_demographic_df["Total Count"] / len(gender_count_df) 
age_demographic_df["Percentage of Players"] = age_demographic_df["Percentage of Players"].astype(float).map("{:,.2%}".format)
age_demographic_df
```
